```python
# File: iris_visualizations_matplotlib.py

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
iris = sns.load_dataset('iris')

# Scatter Plot
plt.figure(figsize=(10, 6))
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=species)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.legend()
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(iris['petal_length'], bins=20, edgecolor='k', alpha=0.7)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()

# Subplots for Sepal Length and Width by Species
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    axes[0].hist(subset['sepal_length'], bins=10, alpha=0.5, label=species)
    axes[1].hist(subset['sepal_width'], bins=10, alpha=0.5, label=species)

axes[0].set_xlabel('Sepal Length')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram of Sepal Length by Species')

axes[1].set_xlabel('Sepal Width')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of Sepal Width by Species')

axes[0].legend()
axes[1].legend()

plt.tight_layout()
plt.show()

# Pairwise Scatter Plots with Subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

axes[0, 0].scatter(iris['sepal_length'], iris['sepal_width'], c=iris['species'].astype('category').cat.codes)
axes[0, 0].set_xlabel('Sepal Length')
axes[0, 0].set_ylabel('Sepal Width')
axes[0, 0].set_title('Sepal Length vs Sepal Width')

axes[0, 1].scatter(iris['petal_length'], iris['petal_width'], c=iris['species'].astype('category').cat.codes)
axes[0, 1].set_xlabel('Petal Length')
axes[0, 1].set_ylabel('Petal Width')
axes[0, 1].set_title('Petal Length vs Petal Width')

axes[1, 0].scatter(iris['sepal_length'], iris['petal_length'], c=iris['species'].astype('category').cat.codes)
axes[1, 0].set_xlabel('Sepal Length')
axes[1, 0].set_ylabel('Petal Length')
axes[1, 0].set_title('Sepal Length vs Petal Length')

axes[1, 1].scatter(iris['sepal_width'], iris['petal_width'], c=iris['species'].astype('category').cat.codes)
axes[1, 1].set_xlabel('Sepal Width')
axes[1, 1].set_ylabel('Petal Width')
axes[1, 1].set_title('Sepal Width vs Petal Width')

plt.tight_layout()
plt.show()
```
